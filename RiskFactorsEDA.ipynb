# Risk Factors and Access to Care EDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
riskdf=pd.read_csv(r"C:\Users\NehaS\Desktop\CS418\RISKFACTORSANDACCESSTOCARE.csv")
riskdf=riskdf.replace(-1111.1,np.NaN)
```

```python
riskdf
```

```python
#Smokers in Illinois
Illinoisdf=riskdf[riskdf['CHSI_State_Name']=='Illinois']
ILsmoke=Illinoisdf[Illinoisdf['Smoker'].notnull()]
ILsmoke=ILsmoke.sort_values(by=['Smoker'])
# width of the bars
barWidth = 2.0
bars = ILsmoke['Smoker']
sns.set(rc={'figure.figsize':(15,10)})
r1=np.arange(len(bars))
yerr=[(abs(2*(x-y))) for x,y in zip(bars,ILsmoke['CI_Min_Smoker'])]
smokerplot=sns.barplot(x=ILsmoke['CHSI_County_Name'],y=bars,ci=yerr,errcolor='black')
smokerplot.set(xlabel='County in Illinois',ylabel='Mean Percentage of Smokers')
plt.xticks(rotation=90)
plt.show()
```

# Insights and Hypotheses about Personal Lifestyle and Health Issues from the visualizations
According to the charts, Counties like Jackson and Du Page appear have healthier habits among the population while counties like Jefferson and Morgan lie on the unhealthier side of almost every chart with higher number of individuals indulging in poor habits. Cook, Knox and Madison appear to lie on the middle range. Comparing this with the insights from the Death and Birth rate charts might reveal patterns for cause and effect. Also, the Demographics of these counties can be studied to determine which counties have higher poverty and how this affects their habits. 

```python
ILUn=Illinoisdf[Illinoisdf['Uninsured'].notnull()]
ILUn.describe()
barWidth = 2.0
bars = ILUn['Uninsured']
bars.describe()
```

```python
Demodf=pd.read_csv(r"C:/Users/NehaS/Desktop/CS418/DEMOGRAPHICS.csv")
Demodf=Demodf[Demodf['CHSI_State_Name']=='Illinois']
cbars=[(x/y)*100 for x,y in zip(bars,Demodf['Population_Size'])]
ILUn['Uninsured']=cbars
ILUn=ILUn.sort_values(by=['Uninsured'])
sns.set(rc={'figure.figsize':(15,10)})
r1=np.arange(len(bars))
smokerplot=sns.barplot(x=ILUn['CHSI_County_Name'],y=ILUn['Uninsured'], palette='winter_r')
smokerplot.set(xlabel='County in Illinois',ylabel='Percentage of individuals uninsured')
plt.xticks(rotation=90)
plt.show()
```

Considering the variation in population amongst different counties, the number of uninsured people is not an accurate estimator. Hence taking the percentage of population from the Demographics CSV will help in painting an accurate picture of the proportion of uninsured individuals by county. 

```python
#Top 5 uninsured counties
labels=['Alexander','Pulaski','Cook','Jackson','Gallatin']
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.pie(ILUn['Uninsured'][-5:],labels=labels[::-1], autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6'])
plt.title("Top 5 uninsured counties as a comparison")
```
