```python
df_leadDeathCauses = pd.read_csv('LEADINGCAUSESOFDEATH.csv')
NanList = [-9999,-2222,-2222.2,-2,-1111,-1111.1,-9998.9]
```

```python
Demo_df = pd.read_csv('DEMOGRAPHICS.csv')
Demo_df_populationSize = Demo_df[['County_FIPS_Code','CHSI_County_Name','CHSI_State_Name','White','Black',
                                  'Hispanic','Native_American','Asian','Population_Size']]
Demo_df_populationSize = Demo_df_populationSize.replace([i for i in NanList],np.NAN)
#print(Demo_df_populationSize.head())
white_population_percentage = Demo_df_populationSize['White']
Total_white_population_arr = [(x*y)/100 for x,y in zip(white_population_percentage,Demo_df_populationSize['Population_Size'])]

black_population_percentage = Demo_df_populationSize['Black']
Total_black_population_arr = [(x*y)/100 for x,y in zip(black_population_percentage,Demo_df_populationSize['Population_Size'])]

hisp_population_percentage = Demo_df_populationSize['Hispanic']
Total_Hispanic_population_arr = [(x*y)/100 for x,y in zip(hisp_population_percentage,Demo_df_populationSize['Population_Size'])]

NativeAmer_population_percentage = Demo_df_populationSize['Native_American']
Asian_population_percentage = Demo_df_populationSize['Asian']

OtherPopulation_percentage = [(x+y) for x,y in zip(Asian_population_percentage,NativeAmer_population_percentage)]
Other_population_arr = [(x*y)/100 for x,y in zip(OtherPopulation_percentage,Demo_df_populationSize['Population_Size'])]

Demo_df_populationSize['White Population'] = pd.Series(Total_white_population_arr,index=Demo_df_populationSize.index)
Demo_df_populationSize['Black Population'] = pd.Series(Total_black_population_arr,index=Demo_df_populationSize.index)
Demo_df_populationSize['Hispanic Population'] = pd.Series(Total_Hispanic_population_arr,index=Demo_df_populationSize.index)
Demo_df_populationSize['Other Population'] = pd.Series(Other_population_arr,index=Demo_df_populationSize.index)
Demo_df_populationSize
```

```python
#Calculation of Percentage of Total Population size affected with Cancer in each county
#total population affected by cancer 
#percentage of population affected by Cancer
#Smoker Cancer population and non smoker cancer population
white_population_Cancer = [((b+c+d+e+f)*y)/100 for b,c,d,e,f,y in zip(final_merged_df['B_Wh_Cancer'],final_merged_df['C_Wh_Cancer'],final_merged_df['D_Wh_Cancer'],final_merged_df['E_Wh_Cancer'],final_merged_df['F_Wh_Cancer'],final_merged_df['White Population'])]
white_population_Cancer

black_population_Cancer = [((b+c+d+e+f)*y)/100 for b,c,d,e,f,y in zip(final_merged_df['B_Bl_Cancer'],final_merged_df['C_Bl_Cancer'],final_merged_df['D_Bl_Cancer'],final_merged_df['E_Bl_Cancer'],final_merged_df['F_Bl_Cancer'],final_merged_df['Black Population'])]
black_population_Cancer

hispanic_population_Cancer = [((b+c+d+e+f)*y)/100 for b,c,d,e,f,y in zip(final_merged_df['B_Hi_Cancer'],final_merged_df['C_Hi_Cancer'],final_merged_df['D_Hi_Cancer'],final_merged_df['E_Hi_Cancer'],final_merged_df['F_Hi_Cancer'],final_merged_df['Hispanic Population'])]
hispanic_population_Cancer

other_population_Cancer = [((b+c+d+e+f)*y)/100 for b,c,d,e,f,y in zip(final_merged_df['B_Ot_Cancer'],final_merged_df['C_Ot_Cancer'],final_merged_df['D_Ot_Cancer'],final_merged_df['E_Ot_Cancer'],final_merged_df['F_Ot_Cancer'],final_merged_df['Other Population'])]
other_population_Cancer

final_merged_df['White_Population_Cancer'] = pd.Series(white_population_Cancer,index=final_merged_df.index)
final_merged_df['Black_Population_Cancer'] = pd.Series(black_population_Cancer,index=final_merged_df.index)
final_merged_df['Hispanic_Population_Cancer'] = pd.Series(hispanic_population_Cancer,index=final_merged_df.index)
final_merged_df['Other_Population_Cancer'] = pd.Series(other_population_Cancer,index=final_merged_df.index)


tot_population_cancer = [((a+b+c+d)/e)*100 for a,b,c,d,e in zip(final_merged_df['White_Population_Cancer'],final_merged_df['Other_Population_Cancer'],final_merged_df['Black_Population_Cancer'],final_merged_df['Hispanic_Population_Cancer'],final_merged_df['Population_Size'])]
final_merged_df['Percentage_Total_Cancer_Population'] = pd.Series(tot_population_cancer,index = final_merged_df.index)
#print(final_merged_df.head(1))
final_merged_illinois = final_merged_df[final_merged_df['CHSI_State_Name_x']=='Illinois']
#final_merged_illinois = final_merged_illinois[['CHSI_County_Name_x','Smoker','Percentage_Total_Cancer_Population']].copy()

corrs = (final_merged_df[['Smoker', 'CHSI_State_Name_x']]
        .groupby('CHSI_State_Name_x')
        .corrwith(final_merged_df.Percentage_Total_Cancer_Population)
        .rename(columns={'Smoker' : 'Corr_Coef'}))
corrs_smoke = corrs[corrs['Corr_Coef'].notnull()]
print(corrs_smoke)
sns.pairplot(corrs_smoke)
```
